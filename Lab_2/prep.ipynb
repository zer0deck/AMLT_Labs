```python
import pandas as pd
import numpy as np

df = pd.read_csv("winequality-red.csv", sep=",")
df.head(5)
```

```python
df['Quality'] = np.where(df.quality > 5, 'good', 'bad')
df = df[['Quality', 'fixed acidity', 'volatile acidity', 'citric acid', 'residual sugar', 'chlorides', 'free sulfur dioxide', 'total sulfur dioxide', 'density', 'pH', 'sulphates', 'alcohol']]
df.head(5)
```

```python
df.to_csv('wine.csv', sep=',', index=None)
```
